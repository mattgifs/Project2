```python
import pandas as pd
import numpy as np
from pathlib import Path
import hvplot.pandas
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from pandas.tseries.offsets import DateOffset
from sklearn.metrics import classification_report
```

```python
# Read the csv file 
amzn_st_path = Path("./Search_Trends/Amazon.csv")
amzn_st = pd.read_csv(amzn_st_path, index_col="Month", infer_datetime_format=True, parse_dates=True)

amzn_st.head()
```

```python
#Plot the line graph
amzn_st.plot(figsize = (20,10))
```

```python
# Read the csv file
meta_st_path = Path("./Search_Trends/Meta.csv")
meta_st = pd.read_csv(meta_st_path, index_col="Month", infer_datetime_format=True, parse_dates=True)

meta_st.head()
```

```python
#Plot the line graph
meta_st.plot(figsize = (20,10), color='orange')
```

```python
# Read the csv file
tsla_st_path = Path("./Search_Trends/Tesla.csv")
tsla_st = pd.read_csv(tsla_st_path, index_col="Month", infer_datetime_format=True, parse_dates=True)

tsla_st.head()
```

```python
#Plot the line graph
tsla_st.plot(figsize = (20,10), color='green')
```

```python
#combined the dataframes
search_trends = pd.concat([amzn_st, meta_st, tsla_st], axis= 1, join = "inner")
```

```python
#visualize the serach trends 
search_trends.plot(figsize = (20,10))
```
